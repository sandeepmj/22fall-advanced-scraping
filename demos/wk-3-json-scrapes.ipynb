```python
## base url for deep dive inspections

base_url = "https://a816-health.nyc.gov/ABCEatsRestaurants/App/GetEntityDetail?camisId={}"
```

```python
## all pages

boros = ["Bronx", "Queens", "Manhattan", "Brooklyn", "Staten%20Island"]
all_df = []
base_url = "https://a816-health.nyc.gov/ABCEatsRestaurants/App/GetEntitiesByBoro/{}"

for boro in boros:
    response = requests.get(base_url.format(boro))
    restaurant_overview = response.json()
    df = pd.DataFrame(restaurant_overview)
    all_df.append(df)
```

```python
all_df[1]
```

```python
def combine_tables(list_name, file_name):
    '''
    takes list of dfs, combines into single csv
    requires csv filename as string
    '''
    df = pd.concat(list_name)
    df.reset_index(inplace = True)
    df.to_csv(file_name, encoding = "UTF-8", index = False)
    return df
```

```python
final_df = combine_tables(all_df, "restaurant-overview.csv")
final_df.sample(20)
```
